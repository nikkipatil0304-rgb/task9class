```python
class TourPlanner:
    def __init__(self, company_name):
        self.company_name = company_name
        self.tour_packages = [
            {"Nikita": "Goa Tour", "price": 200},
            {"Nikita2": "Manali Tour", "price": 300},
            {"Nikita3": "Kashmir Tour", "price": 400}
        ]

    def book(self, person_name, number_of_people, package_name):
        for package in self.tour_packages:
            if package[person_name] == package_name:
                total_bill = package["price"] * number_of_people

                print("Company Name:", self.company_name)
                print("Customer Name:", person_name)
                print("Package Booked:", package_name)
                print("Number of People:", number_of_people)
                print("Total Bill: ₹", total_bill)
                return

        print("Package not found!")
```

```python
planner = TourPlanner("Happy Travels")

planner.book("Nikita", 3, "Goa Tour")
```

```python
planner = TourPlanner("Manali Tour")

planner.book("Nikita2", 4, "Manali Tour")
```
